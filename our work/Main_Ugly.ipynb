```python
import matplotlib.pyplot as plt
import matplotlib
from deepface import DeepFace
```

```python
from numpy import savetxt
from numpy import loadtxt
```

```python
md = "GBU-Dataset/"
```

```python
import os
# create arrays to have the images inside them
bad_q = os.listdir("GBU-Dataset/Bad_Query_Normailized_128x128_gbuReduced")
bad_t = os.listdir("GBU-Dataset/Bad_Target_Normailized_128x128_gbuReduced")
good_q = os.listdir("GBU-Dataset/Good_Query_Normailized_128x128_gbuReduced")
good_t = os.listdir("GBU-Dataset/Good_Target_Normailized_128x128_gbuReduced")
ugly_q = os.listdir("GBU-Dataset/Ugly_Query_Normailized_128x128_gbuReduced")
ugly_t = os.listdir("GBU-Dataset/Ugly_Target_Normailized_128x128_gbuReduced")

#['Bad_Query_Normailized_128x128_gbuReduced', 'Bad_Target_Normailized_128x128_gbuReduced', 'Good_Target_Normailized_128x128_gbuReduced', 'Ugly_Query_Normailized_128x128_gbuReduced', 'Good_Query_Normailized_128x128_gbuReduced', 'Ugly_Target_Normailized_128x128_gbuReduced']

#print(os.listdir("drive/MyDrive/PSU work/GBU-Dataset/Bad_Query_Normailized_128x128_gbuReduced"))
```

```python
# OpenFace + rcFace + SFace (all )
#print(len(emb))
import numpy as np
from numpy import savetxt


for i in ugly_q:
    emb = DeepFace.represent(img_path = md+"Ugly_Query_Normailized_128x128_gbuReduced/"+i, model_name = 'OpenFace',enforce_detection=False)
    npar = np.array(emb)
    savetxt("GBU-Dataset/OpenFace/Ugly_Query/"+i+".csv", npar, delimiter=',')
    
for i in ugly_t:
    emb = DeepFace.represent(img_path = md+"Ugly_Target_Normailized_128x128_gbuReduced/"+i, model_name = 'OpenFace',enforce_detection=False)
    npar = np.array(emb)
    savetxt("GBU-Dataset/OpenFace/Ugly_Target/"+i+".csv", npar, delimiter=',')
    
```

```python
#IMP
ugly_qe = os.listdir("GBU-Dataset/OpenFace/Ugly_Query/")
ugly_te = os.listdir("GBU-Dataset/OpenFace/Ugly_Target/")
#data = loadtxt("GBU-Dataset/Bad_Query/"+i, delimiter=',')

diff = []
don = []

for i in ugly_qe:
    idq = i.split("d")[0]
    if idq not in don:
        embq = loadtxt("GBU-Dataset/OpenFace/Ugly_Query/"+i, delimiter=',')
        for j in ugly_te:
            idt = j.split("d")[0]
            embt = loadtxt("GBU-Dataset/OpenFace/Ugly_Target/"+j, delimiter=',')
            if idq != idt:
                diff.append(DeepFace.dst.findCosineDistance(embq,embt))
    don.append(idq)
    
```

```python
#GIN
ugly_qe = os.listdir("GBU-Dataset/OpenFace/Ugly_Query/")
ugly_te = os.listdir("GBU-Dataset/OpenFace/Ugly_Target/")
#data = loadtxt("GBU-Dataset/Bad_Query/"+i, delimiter=',')
sm = []

don = []

for i in ugly_qe:
    idq = i.split("d")[0]
    if idq not in don:
        embq = loadtxt("GBU-Dataset/OpenFace/Ugly_Query/"+i, delimiter=',')
        for j in ugly_te:
            idt = j.split("d")[0]
            embt = loadtxt("GBU-Dataset/OpenFace/Ugly_Target/"+j, delimiter=',')
            if idq == idt:
                sm.append(DeepFace.dst.findCosineDistance(embq,embt))
    don.append(idq)
    
```

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.kdeplot(diff, bw=0.5)
sns.kdeplot(sm, bw=0.5)
# print(len(diff))
print(sum(diff)/len(diff))
print(sum(sm)/len(sm))
print(len(diff))
print(len(sm))
```

```python
# saving the distinces 
ar1 = np.array(diff)
savetxt("GBU-Dataset/dis/"+"imp_ugly+OpenFace"+".csv", ar1, delimiter=',')
ar2 = np.array(sm)
savetxt("GBU-Dataset/dis/"+"gin_ugly+OpenFace"+".csv", ar2, delimiter=',')
```
